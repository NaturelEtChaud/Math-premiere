1) Donner la valeur demandée après exécution du programme.

```python
def mystere1(n):
    s = 0
    for i in range(n):
        s = s + i
    return s
```

```python
mystere1(5)
```

**Aide:** Pour comprendre ce qui se passe à chaque boucle, on peut rajouter la fonction `print()`. <br>
Il ne faut pas oublier non plus que la première valeur de `i` est 0, donc les valeurs prises par `i` sont succéssivement 0, 1, 2, ..., $n-1$.

```python
def mystere1(n):
    s = 0
    for i in range(n):
        s = s + i
        print(s)
    return s
```

```python
mystere1(5)
```

3) Donner la valeur demandée après exécution du programme.

```python
def mystere3(n):
    s = 0
    for i in range(n):
        s = s + (-i)**i
    return s
```

```python
mystere3(4)
```

**Aide :** En rajoutant la fonction `print()` et en se rappelant que le signe de `(-i)**i` dépend de la puissance paire ou impaire.

```python
def mystere3(n):
    s = 0
    for i in range(n):
        s = s + (-i)**i
        print(i)
    return s
```

```python
mystere3(4)
```

4) Donner la valeur demandée après exécution du programme.

```python
def mystere4(n):
    s = 0
    for i in range(n):
        s = s + i
        if s == 6:
            s = s/2
    return s
```

```python
mystere4(5)
```

**Aide :** En rajoutant la fonction `print()` et en se rappelant que `if s == 6:` signifie  *si i est égal à 6 alors ...*

```python
def mystere4(n):
    s = 0
    for i in range(n):
        s = s + i
        print(s)
        if s == 6:
            s = s/2
            print(s)
    return s
```

```python
mystere4(5)
```

5) Donner la valeur demandée après exécution du programme.

```python
def mystere5(n):
    s = 0
    for i in range(n):
        s = s + i**2
        if s == 3 :
            s = s + 1
        else :
            s = s - 1
    return s
```

```python
mystere5(4)
```

**Aide :** En rajoutant la fonction `print()` et en se rappelant que `else :` signifie  *sinon ...*

```python
def mystere5(n):
    s = 0
    for i in range(n):
        s = s + i**2
        print(s)
        if s == 3 :
            s = s + 1
        else :
            s = s - 1
        print(s)
    return s
```

```python
mystere5(4)
```

6) Donner la valeur demandée après exécution du programme.

```python
def mystere6(n):
    s = 0
    for i in range(n):
        s = s + (-i)**i
        if s < 0:
            s = -s
        else :
            s = s + 2
    return s
```

```python
mystere6(4)
```

**Aide :** En rajoutant la fonction `print()`.

```python
def mystere6(n):
    s = 0
    for i in range(n):
        s = s + (-i)**i
        print(s)
        if s < 0:
            s = -s
        else :
            s = s + 2
        print(s)
    return s
```

```python
mystere6(4)
```
