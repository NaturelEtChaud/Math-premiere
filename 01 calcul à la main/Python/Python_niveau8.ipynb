4) Que renvoie `mystere(5)` et `mystere(13)`?

```python
def mystere(a):
    while a < 13 :
        a = a + 3
    return a
```

```python
mystere(5)
```

```python
mystere(13)
```

**Aide :** En rajoutant la fonction `print()` on peut voir l'évolution de la variable `a` à chaque boucle.

```python
def mystere(a):
    while a < 13 :
        a = a + 3
        print(a)
    return a
```

```python
mystere(5)
```
