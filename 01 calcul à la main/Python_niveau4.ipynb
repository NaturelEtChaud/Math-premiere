### Niveau 4 : Donner la valeur demandée après exécution du programme

```python
def f(x):
    return 3*x + 4

f(-5)
```

---

```python
def g(t):
    return t**2 + 3*t - 4

g(4)
```

---

```python
def aire(larg, longu):
    return larg * longu

aire(3,8)
```

---

```python
def f(a, b):
    return a*b - a - b

f(4, 3)
```

---

```python
def peri(l, L):
    return 2 * (l + L)

peri(12, 3)
```

---

```python
def aire_tri(b, h):
    return b * h / 2

aire_tri(10,8)
```
