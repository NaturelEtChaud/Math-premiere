### Niveau 5 : Donner la valeur demandée après exécution du programme
---

```python
def f(a):
    if a < 0:
        b = 2*a
    else :
        b = 3*a 
    return b
```

```python
f(1)
```

```python
f(-1)
```

```python
def f(a):
    if a >= 1:
        b = a - 1
    else :
        b = 2*a 
    return b
```

```python
f(1)
```

```python
f(-1)
```

---

```python
def f(a, b):
    if a > b:
        y = a - b
    else :
        y = b - a 
    return y
```

```python
f(4, 7)
```

```python
f(7, 4)
```

```python
def f(t):
    if t < 2:
        calcul = 2*t
    else :
        calcul = 2*(t - 2) + 1
    return calcul
```

```python
f(2)
```

```python
f(-2)
```
